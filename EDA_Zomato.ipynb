```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:"Rating_Counts"})
```

## Observations
1. When ratings are in between 5.0 to 4.5 ---> Excellent
2. When ratings are in between 4.0 to 4.4 ---> Very Good
3. When ratings are in between 3.5 to 3.9 ---> Good
4. When ratings are in between 2.5 to 3.5 ---> Average
5. When ratings are in between 1.8 to 2.5 ---> poor

```python
ratings.head()
```

```python
import matplotlib

```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating',y ='Rating_Counts',data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating',y ='Rating_Counts',data = ratings,hue ='Rating color')
```

```python
sns.barplot(x = 'Aggregate rating',y ='Rating_Counts',data = ratings,hue ='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

## Observations

1. Not rated has the highest number of instances
2. Most rated values lie in betweeen 2.5 to 3.4
